```python
df4 = pd.read_csv('/home/simon/Data/mri-augmentation/spreadsheets/Torsion/torsion_table_long_format_v2.csv')
```

```python
df4.head()
```

```python
df4['side'] = df4['side'].apply(lambda x: 'A_right' if x == 'right' else x)
df4['severity'] = df4['severity'].apply(lambda x: 'A_reference' if x == 'reference' else x)
df4['type'] = df4['type'].apply(lambda x: 'A_femur' if x == 'femur' else x)
df4['reader'] = df4['reader'].apply(lambda x: 'A_baseline' if x == 'baseline' else x)
df4['reader'] = df4['reader'].apply(lambda x: 'B_Default' if x == 'default' else x)
df4['reader'] = df4['reader'].apply(lambda x: 'C_MRI' if x == 'mr' else x)
df4['reader'] = df4['reader'].apply(lambda x: 'D_Marc' if x == 'marc' else x)
df4['reader'] = df4['reader'].apply(lambda x: 'E_Robert' if x == 'robert' else x)
```

```python
df4[['type', 'side', 'severity', 'reader', 'torsion']].groupby(by=['type', 'side', 'severity', 'reader']).agg({'torsion': ['mean', 'std']})
```
